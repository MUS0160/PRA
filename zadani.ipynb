### Aktivita genu

V pravidelných intervalech sledujeme aktivitu určitého genu v organismu. Aktivita se může v průběhu času měnit.
Předpokládejme, že v každém okamžiku $t\in\{1, 2, \ldots, T\}$ je gen buď aktivní, nebo neaktivní. Průběh aktivity genu v čase můžeme modelovat jako posloupnost náhodnýh veličin $X_1, X_2, \ldots, X_T$, kde $X_t=1$, pokud je gen v čase $t$ aktivní a $X_t=0$, jestliže je gen v čase $t$ neaktivní. Je-li gen aktivní, zvyšuje se v organismu produkce určitého proteinu. Pokud je neaktivní, je protein organismem také produkován, ale v menším množství. Protože se protein v organismu rychle rozkládá, je jeho množství během měření závislé pouze na aktivitě genu v příslušném okamžiku.

Předpokládejme, že aktivity genu $X_1, X_2, \ldots, X_T$ nejsou nezávislé náhodné veličiny, ale splňují následující vlastnost: pro každý čas $t$ platí, že náhodná veličina $X_t$ je podmíněně nezávislá na $X_1, X_2, \ldots, X_{t-2}$ za podmínky $X_{t-1}$. Jinak řečeno, platí 
\begin{equation}
\forall t\in\{2, 3, \ldots, T\}, \forall (x_1, \ldots, x_t)\in\mathbb \{0, 1\}^t: (X_t=x_t) \mathrel{\unicode{x2AEB}} (X_1=x_1 \cap\cdots\cap X_{t-2}=x_{t-2}) | (X_{t-1}=x_{t-1}),
\end{equation} 
a tedy 
\begin{equation}
P(X_t=x_t|X_1=x_1 \cap\cdots\cap X_{t-1}=x_{t-1})=P(X_t=x_t|X_{t-1}=x_{t-1}).
\end{equation} 
Dále předpokládejme, že pravděpodobnosti změn aktivity genu se v čase nemění, tj. $P(X_t=j|X_{t-1}=i)$ nezávisí na čase $t$. 
Pro každé $(i, j)\in\{0,1\}^2$ Označme
\begin{equation}
Q_{ij}:=p_{X_t|X_{t-1}}(j, i)=P(X_t=j|X_{t-1}=i).
\end{equation} 
Hodnoty $Q_{ij}$ vyjadřují pravděpodobnosti, že aktivita genu v následujícím čase bude rovna hodnotě $j$, je-li její hodnota v aktuálním čase rovna $i$. Pravděpodobnosti $Q_{ij}$ nazýváme pravděpodobnosnti přechodu a můžeme je uspořádat do tzv. přechodové matice
\begin{equation}
Q = \left(Q_{ij}\right)_{i,j=0,0}^{1,1}.
\end{equation}
Předpokládejme, že pravděpodobnosti přechodu odpovídají přechodové matici 
        \begin{equation}
        Q=\left(\begin{array}{ll}  0.8 & 0.2\\ 0.1 & 0.9\end{array}\right).
        \end{equation}
